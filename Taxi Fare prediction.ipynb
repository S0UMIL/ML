```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv('taxi_trip_pricing.csv')
data = data.dropna()
distance=data['Trip_Distance_km'].values
cost=data['Trip_Price'].values
```

```python
plt.scatter(distance,cost,marker='o',color='blue')
plt.title('Taxi Fares VS Distance')
plt.xlabel('Distance covered')
plt.ylabel('Cost')
plt.show()
```

```python
x_train,x_test,y_train,y_test=train_test_split(distance,cost,test_size=0.2,random_state=50)
```

```python
print(np.isnan(distance).sum())
print(np.isnan(cost).sum())
```

```python
x_train=x_train.reshape(-1,1)
x_test=x_test.reshape(-1,1)
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
predictions=model.predict(x_test)
plt.scatter(x_test,y_test,marker='o',label='Actual Cost')
plt.plot(x_test,predictions,color='red',linewidth=2,label='Predicted Cost')
plt.title('Distance VS taxi prices Prediction')
plt.xlabel('Distance covered')
plt.ylabel('Cost')
plt.legend()
plt.show()
```
