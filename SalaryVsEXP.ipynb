```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv('Salary_dataset.csv')
experience=data['YearsExperience'].values
salary=data['Salary'].values
```

```python
#plotting the graph
plt.scatter(experience,salary,marker='o',color='blue')
plt.title('YearsOfExperience VS Salary')
plt.xlabel('Years of Experience')
plt.ylabel('salary (in $)')
plt.show()
```

```python
x_train,x_test,y_train,y_test=train_test_split(experience,salary,test_size=0.2,random_state=42)
```

```python
x_train=x_train.reshape(-1,1)
x_test=x_test.reshape(-1,1)
model= LinearRegression()
model.fit(x_train,y_train)
```

```python
predictions=model.predict(x_test)
plt.scatter(x_test,y_test,marker='o',color='blue',label='Actual Salaries')
plt.plot(x_test,predictions,color='green',linewidth=2,label='Predicted Salaries')
plt.title('Years of experience vs Salaries with LinearRegression')
plt.xlabel('Years of Experience')
plt.ylabel('salary (in $)')
plt.legend()
plt.show()
```
